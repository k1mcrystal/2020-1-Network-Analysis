````python
# 네트워크 빈 도화지
G = nx.Graph()

G
````

````python
G.add_node(1)

nx.draw(G, with_labels=True)
````

````python
G.add_nodes_from([2, 3])

nx.draw(G, with_labels=True)
````

````python
H = nx.path_graph(10) #일렬로 연결된 n개의 node를 return합니다(순서는 무작위).
nx.draw(H, with_labels = True, node_color = 'coral')
````

````python
G.add_nodes_from(H)

nx.draw(G, with_labels = True, node_color = 'coral')
````

````python
G.add_edge(1, 2)

e = (2, 3)
G.add_edge(*e) # unpack edge tuple*

nx.draw(G,
        with_labels=True,
        node_color="skyblue"
       )
# matplotlib의 color setting과 동일
````

````python
H = nx.path_graph(10)

G.add_edges_from(H.edges)

print(G.number_of_nodes())
print(G.number_of_edges())

nx.draw(G,
        with_labels=True,
        node_color="skyblue"
       )
````

#### 네트워크를 그려봅시다

````python
# 빈 그래프 생성
graph_miseables = nx.Graph()
````

````python
# Node List에서 name 항목만 추출한다.
node_names = [x['name'] for x in nodes]

# 각 항목을 Graph에 Node로 등록한다.
graph_miseables.add_nodes_from(node_names)
````

````python
# Edge List에서 (source_name, target_name) 형태로 추출한다
link_indices = [(node_names[x['source']], node_names[x['target']]) for x in links]

# 각 항목을 Graph에 Edge로 등록한다
graph_miseables.add_edges_from(link_indices)
````

````python
# link_indices
````

````python
# matplotlib 설정
%config InlineBackend.figure_format = 'retina'
plt.figure(figsize=(20, 15)) # 그래프 size
plt.axis('off') # 축 제거
nx.draw_networkx(graph_miseables) # 그래프 그리기
````

위 그래프를 보면, 각 등장인물(node)들이 어떤관계를 가지는지 여부 정도만... 파악할 수 있습니다. (사실 그 것도 눈에 확 들어오진 않아요.)<br>
누가 영향력이 있는지, 누구와 큰 관계를 맺고 있는 지 등을 표시할 필요가 있어 보입니다.<br>

#### Node Weight 반영하기

Node들의 중요도를 파악할 수 있는 지표에는 여러가지가 있습니다.
- 밀도 (네트워크 내 전체 노드간 맺고 있는 관계의 밀집 정도) Density<br>
: degree, inclusiveness<br>
<br>
- 중심성 (한 노드가 전체 네트어크에서 중심에 위치하는 정도) Centrality<br>
: degree, closeness, betweenness<br>
<br>
- 중심화 (네트워크의 형태가 중앙에 집중되어 있는 정도) Centralization<br>


참고 ) https://networkx.github.io/documentation/stable/reference/algorithms/centrality.html

우선 가장 간단한 degree에 대해서 살펴볼까요?<br>
Degree 값을 계산해서 Node 크기에 반영해봅시다. `.degree()` 메서드를 사용하면 각 Node의 degree 값을 구할 수 있습니다.

`networkx.draw_networkx` 함수에서 `node_size` 속성을 변경하면 각 Node의 크기를 조절할 수 있습니다.

````python
mis_degree = dict(graph_miseables.degree())
mis_degree
````

````python
max_value = max(mis_degree.values())  # maximum value
max_keys = [k for k, v in mis_degree.items() if v == max_value] # getting all keys containing the `maximum`

print(max_value, max_keys)
````

````python
# 많은 관계를 가질 수록 node를 크게 만들기
graph_degrees = graph_miseables.degree() # Degree 값 구하기
node_sizes = [graph_degrees[x] * 50 for x in node_names] # Node Size로 Rescale

plt.figure(figsize=(20, 15))
plt.axis('off')
nx.draw_networkx(graph_miseables, node_size=node_sizes, node_color='#88B04B', alpha=0.5)
````

#### Edge Weight 반영하기

Edge Weight (선의 두께)는 `networkx.draw_networkx` 함수에서 `width` 속성을 변경하면 됩니다. 그래프를 생성할 때 넣었던 Edge 순서대로 값을 넣어줍니다.

````python
links
````

````python
[x['value'] for x in links]
````

````python
# 관계가 강할수록 edge의 굵기를 굵게하기
# links에서 value 항목만 추출한다
edge_width = [x['value'] for x in links]
# 1~10 정도로 스케일을 조정한다
edge_width = [(x / max(edge_width) * 9) + 1 for x in edge_width]

plt.figure(figsize=(20, 15))
plt.axis('off')
nx.draw_networkx(graph_miseables, 
                 node_size=node_sizes, width=edge_width,
                 node_color='#FF69B4', alpha=0.5)
````
